```python
from scipy.stats import chi2_contingency
sample_size=10
Loan_sample=np.random.choice(df['ApplicantIncome'],sample_size)
print("The samples are:",Loan_sample)
stat,p,dof,expected=chi2_contingency(Loan_sample)
alpha=0.05
print("p value is"+str(p))
if p<=alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
    
```
